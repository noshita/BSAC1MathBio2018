## 四則演算など

```
# 加算
print("1+1 = ", 1+1)
print("x+1 = ", x+1)
# 減算
print("2-1 = ", 2-1)
print("y-1 = ", y-1)
# 乗算
print("2*5 = ", 2*1)
print("x*y = ", x*y)
# 除算
print("2/5 = ", 2/5)
print("x/y = ", x/y)
# 余り
print("14%5 = ", 14%5)
print("17%x = ", 17%x)
```

## リスト List

```
a = [x, y]
b = [x, y, s, bo]
print(a)
print(b)
```

```
# 要素
print(a[0])
print(b[-1])
```

```
# スライス　slice
print(b[1:3])
print(b[:3])
print(b[1:])
print(b[:])
```
